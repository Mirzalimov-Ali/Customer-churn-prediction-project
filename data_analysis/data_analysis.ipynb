```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('../data/raw/telco_customer_churn.csv')
```

```python
df.info()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Churn', data=df)
plt.title('Distribution of Churn')
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("MonthlyCharges vs Churn")
plt.show()
```

# Observation: The dataset is imbalanced, with customers who did not churn being slightly more than twice as many as those who did churn.

```python
plt.figure(figsize=[6, 6])
plt.hist(df['gender'], bins=2, color='skyblue', edgecolor='black')
plt.show()
```

```python
sns.countplot(x='gender', hue='Churn', data=df)
plt.title("Churn by Gender")
plt.show()
```

# Observation: The gender distribution among customers who churned is roughly 50/50, indicating that gender does not significantly impact churn in this dataset.

```python
plt.figure(figsize=[6, 5])
plt.hist(df['Contract'], bins=3, color='skyblue', edgecolor='black')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title('Contract Type vs Churn')
plt.show()
```

# Observation: Customers with a two-year contract, although more numerous than one-year contracts, have a lower churn rate, meaning they are less likely to leave than customers with shorter contracts.

```python
plt.figure(figsize=(6,4))
sns.countplot(x='InternetService', hue='Churn', data=df, palette='Set1')
plt.title('Internet Service vs Churn')
plt.show()
```

# Observation: Customers using Fiber optic have a much higher churn rate compared to DSL or No internet service.

```python
plt.figure(figsize=(8,4))
sns.histplot(data=df, x='tenure', hue='Churn', multiple='stack', bins=30)
plt.title('Tenure vs Churn')
plt.show()
```

# Observation: Customers with very low tenure (just joined) have a higher churn rate compared to those who stayed longer.

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Partner', hue='Churn', data=df)
plt.title('Partner vs Churn')
plt.show()
```

# Observation: Customers without a partner churn slightly more than those who have a partner.

```python
fig = px.scatter(df, x='tenure', y='MonthlyCharges', color='Churn', hover_data=['gender', 'Contract', 'InternetService'])
fig.show()
```

# Customers who churn frequently have the following characteristics:
    Mostly Fiber optic internet users.
    Mostly Month-to-month contracts.
    Usually new customers (low tenure, 0–2 months).
    Slightly more likely to be without a partner.
    Monthly charges are generally higher than average.
